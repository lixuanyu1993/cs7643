```python
import numpy as np
import numpy as np
from scipy.interpolate import make_interp_spline
import matplotlib.pyplot as plt
```

```python
def parse_init(init_l):
    result = []
    for l in init_l:
        result.append(l[2])
    return result
```

```python
def parse_init_list(init_ll):
    result = []
    for y in init_ll:     
        result.append(parse_init(y))
    return result
```

```python
def smooth(scalars, weight):  
    last = scalars[0]
    smoothed = list()
    for point in scalars:
        smoothed_val = last * weight + (1 - weight) * point  
        smoothed.append(smoothed_val)                        
        last = smoothed_val                                  

    return smoothed
```

```python
def draw(x, y_list_init, name_list,title="", top=0.025, bottom=0.015):
    y_list = parse_init_list(y_list_init)
    plt.grid(True)
    plt.title(" vs ".join(name_list) + title, fontsize=14)
    plt.xlabel('Epochs', fontsize=12)
    plt.ylim(top=0.025, bottom=0.015)
    plt.ylabel('Eval', fontsize=12)
    for i in range(len(y_list)):
        y = y_list[i]
        name = y_name_list[i]
        plt.plot(x, smooth(y, 0.6),linewidth=1,label=name)
    plt.legend()
```
